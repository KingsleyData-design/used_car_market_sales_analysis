```python
import matplotlib.pyplot as plt

avg_price = df.groupby("company")["price"].mean().sort_values(ascending=False)

avg_price.plot(kind="bar", figsize=(14,6), color="steelblue")
plt.title("Average Price by Car Company")
plt.ylabel("Price (INR)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
df["company"].value_counts().plot(kind="bar", figsize=(14,6), color="steelblue")
plt.title("Number of Listings by Company")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
df["fuel_type"].value_counts().plot(kind="pie", figsize=(6,4), color="steelblue")
plt.title("Fuel Type Distribution")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df["company"].value_counts().plot(kind="bar", figsize=(14,6), color="steelblue")
plt.title("Number of Cars by Company")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
df.groupby("year")["price"].mean().plot(kind="line", figsize=(6,4), color="steelblue", marker="o")
plt.title("Average Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Price (INR)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["kms_driven"], df["price"], alpha=0.4, color="steelblue")
plt.title("Price vs Kms Driven")
plt.xlabel("Kms Driven")
plt.ylabel("Price (INR)")
plt.tight_layout()
plt.show()
```
